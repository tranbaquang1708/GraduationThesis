```
import sys
import os
mp = os.path.abspath(os.path.join('/content/drive/My Drive/GraduationThesis/gitrepo/GraduationThesis/'))
sys.path.append(mp)
```

```
from modules import Rbf
```

```
# Off surface points
filename = mp + '/dataset/3d/sphere.xyz'
dataset, normal_vectors = Rbf.gen_from_txt3(filename)
```

```
#Solve weight function
w = Rbf.solve_w_func(dataset, offset=0.01, func=Rbf.linear)
```

```
# Evaluate f(x,y) on a 2D regular grid obtained from meshgrid
xx,yy,zz = Rbf.grid(dataset[:,0], dataset[:,1], dataset[:,2])
z = Rbf.sampling(dataset, w, xx, yy, zz, func=Rbf.linear)
```

```
# Visualize
Rbf.visualize3(dataset, normal_vectors, z)
```
