```
# Generate ssh key
# Note: Change parameters before run (user and file name)

! rm -rf /root/.ssh*
! mkdir /root/.ssh
%cd /root/.ssh

! ssh-keygen -t rsa -b 4096 -C "user@github.com" # substitute user
# Enter file name
# No parapharse

! chmod 700 /root/.ssh
```

```
# Set up connection
# The following lines assume that file name in previous block is "id_rsa"
# Note: Change parameters before run

! cat id_rsa.pub # file name might need to be changed
# Copy file content
# Add as new ssh key on Github

! touch /root/.ssh/known_hosts
! chmod 644 /root/.ssh/known_hosts

! ssh-keyscan github.com >> /root/.ssh/known_hosts

! touch /root/.ssh/config
! echo 'Host github.com' >> /root/.ssh/config
! echo '  Hostname github.com' >> /root/.ssh/config
! echo '  User git' >> /root/.ssh/config
! echo '  IdentityFile /root/.ssh/id_rsa' >> /root/.ssh/config # file name might need to be changed

! git config --global user.email "user@mail.com" # substitute user and mail domain
! git config --global user.name "user" # substitute user
```
